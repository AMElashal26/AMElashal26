## What is DagsHub?
DagsHub is a platform for ML and AI teams that let's them organize and collaborate on data, code, experiments, and models in one place.

It combines open source tools and formats like Git, DVC, MLflow, Label Studio, and more.

DagsHub also contains with [DagsHub Data Engine](https://dagshub.com/data-engine/) â€“Â A tool for versioning datasets, filtering by custom metadata, annotating it, and using it for analysis or training.

## What is DagsHub Storage?
Each project on DagsHub comes with [DagsHub Storage](https://dagshub.com/docs/feature_guide/dagshub_storage/), an S3-compatible bucket, that you can use to upload, download and manage your data.

## Why DagsHub & Colab work better together?
In the Colab ecosystem, DagsHub Storage provides an ML-oriented alternative to GDrive, supporting standard tools like Boto, S3FS, and RClone, as well as mounting to Colab directly. This is in addition to DagsHub functionality for diffing and visualizing notebooks which can be useful

This is especially useful for teams working with large datasets, or that want an easy way to see the data alongside their code, experiments, and more.

### ğŸƒâ€â™€ï¸ğŸƒğŸ»ğŸƒâ€â™‚ï¸Prep for training, just run this, only boilerplate inside

```python
import os
import torch

from torchvision.io import read_image
from torchvision.ops.boxes import masks_to_boxes
from torchvision import tv_tensors
from torchvision.transforms.v2 import functional as F


class PennFudanDataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms):
        self.root = root
        self.transforms = transforms
        # load all image files, sorting them to
        # ensure that they are aligned
        self.imgs = list(sorted(os.listdir(os.path.join(root, "PNGImages"))))
        self.masks = list(sorted(os.listdir(os.path.join(root, "PedMasks"))))

    def __getitem__(self, idx):
        # load images and masks
        img_path = os.path.join(self.root, "PNGImages", self.imgs[idx])
        mask_path = os.path.join(self.root, "PedMasks", self.masks[idx])
        img = read_image(img_path)
        mask = read_image(mask_path)
        # instances are encoded as different colors
        obj_ids = torch.unique(mask)
        # first id is the background, so remove it
        obj_ids = obj_ids[1:]
        num_objs = len(obj_ids)

        # split the color-encoded mask into a set
        # of binary masks
        masks = (mask == obj_ids[:, None, None]).to(dtype=torch.uint8)

        # get bounding box coordinates for each mask
        boxes = masks_to_boxes(masks)

        # there is only one class
        labels = torch.ones((num_objs,), dtype=torch.int64)

        image_id = idx
        area = (boxes[:, 3] - boxes[:, 1]) * (boxes[:, 2] - boxes[:, 0])
        # suppose all instances are not crowd
        iscrowd = torch.zeros((num_objs,), dtype=torch.int64)

        # Wrap sample and targets into torchvision tv_tensors:
        img = tv_tensors.Image(img)

        target = {}
        target["boxes"] = tv_tensors.BoundingBoxes(boxes, format="XYXY", canvas_size=F.get_size(img))
        target["masks"] = tv_tensors.Mask(masks)
        target["labels"] = labels
        target["image_id"] = image_id
        target["area"] = area
        target["iscrowd"] = iscrowd

        if self.transforms is not None:
            img, target = self.transforms(img, target)

        return img, target

    def __len__(self):
        return len(self.imgs)

import torchvision
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from torchvision.models.detection.mask_rcnn import MaskRCNNPredictor


def get_model_instance_segmentation(num_classes):
    # load an instance segmentation model pre-trained on COCO
    model = torchvision.models.detection.maskrcnn_resnet50_fpn(weights="DEFAULT")

    # get number of input features for the classifier
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    # replace the pre-trained head with a new one
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

    # now get the number of input features for the mask classifier
    in_features_mask = model.roi_heads.mask_predictor.conv5_mask.in_channels
    hidden_layer = 256
    # and replace the mask predictor with a new one
    model.roi_heads.mask_predictor = MaskRCNNPredictor(
        in_features_mask,
        hidden_layer,
        num_classes
    )

    return model

from torchvision.transforms import v2 as T


def get_transform(train):
    transforms = []
    if train:
        transforms.append(T.RandomHorizontalFlip(0.5))
    transforms.append(T.ToDtype(torch.float, scale=True))
    transforms.append(T.ToPureTensor())
    return T.Compose(transforms)
```

### Actual training section, note we are loading the dataset from the mount!

```python
from engine import train_one_epoch, evaluate
import utils
import dagshub
import mlflow

dagshub.init(repo_owner=DAGSHUB_REPO.split("/")[0], repo_name=DAGSHUB_REPO.split("/")[1], mlflow=True)

# train on the GPU or on the CPU, if a GPU is not available
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

# our dataset has two classes only - background and person
num_classes = 2

# use our MOUNTED dataset and defined transformations
dataset = PennFudanDataset(f'{mount_path}/data/PennFudanPed', get_transform(train=True))
dataset_test = PennFudanDataset(f'{mount_path}/data/PennFudanPed', get_transform(train=False))

# split the dataset in train and test set
indices = torch.randperm(len(dataset)).tolist()
dataset = torch.utils.data.Subset(dataset, indices[:-50])
dataset_test = torch.utils.data.Subset(dataset_test, indices[-50:])

# define training and validation data loaders
data_loader = torch.utils.data.DataLoader(dataset, batch_size=2, shuffle=True, num_workers=2, collate_fn=utils.collate_fn)

data_loader_test = torch.utils.data.DataLoader(dataset_test, batch_size=1, shuffle=False, num_workers=2, collate_fn=utils.collate_fn)

# get the model using our helper function
model = get_model_instance_segmentation(num_classes)

# move model to the right device
model.to(device)

with mlflow.start_run():
  # construct an optimizer
  params = [p for p in model.parameters() if p.requires_grad]
  optimizer = torch.optim.SGD(params, lr=0.005, momentum=0.9, weight_decay=0.0005)

  # and a learning rate scheduler
  lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)

  # let's train it just for 2 epochs
  num_epochs = 2

  mlflow.log_params({"initial_lr": optimizer.param_groups[0]["lr"],
                     "momentum": optimizer.param_groups[0]["momentum"],
                     "weight_decay": optimizer.param_groups[0]["weight_decay"],
                     "epochs": num_epochs})

  step_counter = 0

  for epoch in range(num_epochs):
      # train for one epoch, printing every 10 iterations
      metric_logger, step_counter = train_one_epoch(model, optimizer, data_loader, device, epoch, print_freq=10, step_counter=step_counter)
      # update the learning rate
      lr_scheduler.step()
      # evaluate on the test dataset
      evaluator = evaluate(model, data_loader_test, device=device)

  mlflow.pytorch.log_model(model, "model")

  print("That's it!")
```

Configure the RClone conf (this is done automatically if using `dagshub.storage.rclone.rclone_init(repo_owner, conf_path = None, update=False, quiet=False)`

### RClone Configuration

```python
import configparser
import os
from pathlib import Path

def rclone_cfg(repo_owner: str, conf_path: Path = None, update=False) -> Path:
    if conf_path is None:
      root_path = Path.home() / ".config/rclone/"
      conf_path = root_path / "rclone.conf"
      if not os.path.exists(root_path):
        os.makedirs(Path.home() / ".config/rclone/")
    else:
      conf_path = Path(conf_path)

    token = dagshub.auth.get_token()

    config = configparser.ConfigParser()
    if conf_path.exists():
        config.read(conf_path)

    # Determine the section name
    if "dagshub" in config and not update:
        section_name = f"dagshub-{repo_owner}"
    else:
        section_name = "dagshub"

    config[section_name] = {
        "type": "s3",
        "provider": "Other",
        "access_key_id": token,
        "secret_access_key": token,
        "endpoint": f"https://dagshub.com/api/v1/repo-buckets/s3/{repo_owner}",
    }

    with conf_path.open("w") as f:
        config.write(f)

    # Inform the user about the remote name
    print(f"Configuration complete. The remote '{section_name}' has been created/updated in '{conf_path}'.")
    print(f"Example usage with rclone: `rclone ls {section_name}:<your-bucket-name>` to list the contents of 'your-bucket-name'.")

    return conf_path.absolute()
```

## Uploading from Python using boto3

```python
%pip install -q -U boto3
```

```python
LOCAL_PATH = "/content/sample_data"
PATH_IN_BUCKET = "sample_data_boto"

from dagshub import get_repo_bucket_client
# Get a boto3.client object
s3 = get_repo_bucket_client(f"{DAGSHUB_REPO_OWNER}/{DAGSHUB_REPO_NAME}")
```

```python
# Upload file
s3.upload_file(
    Bucket=BUCKET,  # name of the repo
    Filename=f"{LOCAL_PATH}/mnist_train_small.csv",  # local path of file to upload
    Key=f"{PATH_IN_BUCKET}/mnist_train_small.csv",  # remote path where to upload the file
)
# Download file
s3.download_file(
    Bucket=BUCKET,  # name of the repo
    Key=f"{PATH_IN_BUCKET}/mnist_train_small.csv",  #  remote path from where to download the file
    Filename="downloaded_mnist_train_small.csv",  # local path where to download the file
)
```
